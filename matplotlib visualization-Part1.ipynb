```python
#this is for matplot basics
import matplotlib.pyplot as plt
%matplotlib inline 
#this is for showing the pic
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
#1.functional method
plt.plot(x,y,'r-')
plt.xlabel('x Label')
plt.ylabel('y Label')
plt.title('Title')
```

```python
#1.2.multiplot 即不止一条线
plt.subplot(1,2,1) #what does this mean ?
plt.plot(x,y,'r-')

plt.subplot(1,2,2)
plt.plot(y,x,'b-')
```

```python
#2. object-oriented method
fig = plt.figure() #create a fig object
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
#2.1 multi figure
fig = plt.figure() #create a fig object
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])  #这些是percent 
#create an inserted axes
axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
#总结：matplotlib.plot有两种画图方式，一种是用函数，一种是创建fig obj
```

```python
#*************************MatplotBasics: Part Two *********************
```

```python
#create two objs 
fig,axes = plt.subplots() # subplots() here functions as axes.set_xlabel()
axes.plot(x,y)
```

```python
#change some parameters in subplot()
fig,axes = plt.subplots(nrows=2,ncols=2)  #这里有4个图，所以有4个对象
#here nrows and ncols decides what kind of list axes is. since nrows=2,ncols=2,its a 2Dlist；if nrows=1,its 1 d list
#so we can access the value by axes[1]
axes = axes.flatten()
#here axes is a list of objs of matplot axes, so its iterable
for current_ax in axes:
    current_ax.plot(x,y)
#also, we can use index to access the value in the list
axes[0].plot(x,y)
axes[0]. set_title('First Plot')
axes[1].plot(y,x)
axes[1]. set_title('Second Plot')

plt.tight_layout() #solve the problem of overlapping

#if its a two dimentional list, we need to use two indexs to access the value, such as axes[0,1]

#so here it shows us how to make several pics in a simple way
```

```python
#*************************** Figure Size and DPI ****************************
```

```python
fig = plt.figure(figsize=(5,2),dpi=100) #dpi means pixles per inch
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes= plt.subplots(nrows=2,ncols=1,figsize=(3,2)) #dpi means pixles per inch
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#legend
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')
ax.legend(loc='upper left') #loc= is used to set the location of the legend

ax.legend(loc=(0.1,0.1)) #you can give it a more detailed location by giving it tuple
#if you use legend(),you must specify the label,but you can use another way
#可以在网上找到所有的location code
```

```python
#****************************Matplotlib Basic-Part Three*******************
#set the color
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red')
#or we can set more customerized color by using #copy from the notebook
ax.plot(x,y,color='#FF8C00')
```

```python
#linewidth and linestyle
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])#
ax.plot(x,y,color='purple',linewidth=3) #can also use lw = 3
#alpha is transparency
ax.plot(x,y,color='purple',linewidth=3,alpha=0.5)
#linestyle
ax.plot(x,y,color='purple',lw=3,linestyle=':') #点图
ax.plot(x,y,color='purple',lw=3,linestyle='steps') 
#or use ls to represent linestyle
```

```python
# marker
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])#
#ax.plot(x,y,color='purple',lw=3,marker='*') #markered 11 points, so why 11 points?
#mark by number codes
#ax.plot(x,y,color='purple',lw=1,marker='1') 
#add marker size,to make the marker larger or smaller
# change marker color
ax.plot(x,y,color='purple',lw=1,marker='o',markersize=10,markerfacecolor='yellow',markeredgecolor='green') 
```

```python
#**************control over axis appearance ***************
#plot range
# what is plot range ? it means you just want to show part of the line
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=2)
ax.set_xlim([0,1]) #xlim mean x limits, so it will only show the plot from 0 to 1
ax.set_ylim([0,2])
```
