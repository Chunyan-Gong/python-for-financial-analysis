```python
import pandas_datareader as pdr
import datetime
import pandas_datareader.data as web
import pandas as pd
from pymongo import MongoClient
from pymongo import UpdateOne
DB_CONN=MongoClient('localhost',27017)
test_db=DB_CONN.test_db
test_db
```

```python
tesla_col=test_db['tesla_col']
tesla_col
```

```python
start = datetime.datetime(2019,8,1)
end = datetime.datetime(2019,8,21)
tesla = web.DataReader('TSLA','yahoo',start,end)
tesla=tesla.reset_index()
type(tesla['Date'][0])
```

```python
for df_index in tesla.index:
    doc = dict(tesla.loc[df_index])
    doc['code'] = 'tesla'
    updaterequest=[]
    updaterequest.append(UpdateOne({'code': doc['code']},
                    {'$set': doc},
                    upsert=True) 
            )
if len(updaterequest)>0:
    update_result=test_db.tesla_col.bulk_write(updaterequest,ordered=False)
update_result
```
