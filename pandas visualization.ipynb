```python
import pandas as pd
import numpy as np
from numpy.random import randn
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#*************************************Pandas Basic Visualization*****************************
```

```python
#1. plot hist
#there are 3 similar approaches
```

```python
df=pd.DataFrame(randn(5,4),[0,1,2,3,4],['A','B','C','D'])
df
```

```python
df['A'].hist()
```

```python
df['B'].plot(kind='hist',bins=30)
```

```python
df['C'].plot.hist()
```

```python
#2. plot area 
df.plot.area() #alpha表示透明度
```

```python
#3. plot.bar()
df.plot.bar()
```

```python
#4.plot line
df.plot.line(x=df.index,y='B',figsize=(12,3))
```

```python
#5. 散点图scatter
df.plot.scatter(x='A',y='B',c='C',cmap='coolwarm') #scatter是很有用的，统计一开始都要做scatter,c表示color
```

```python
df.plot.scatter(x='A',y='B',s=df['C']*100)  #哇 这个超级酷诶，同时可以显示A和B的相关性，然后显示C的value大小
```

```python
#6. box， 显示一组数据中的中值，最大，最小和4分位值，而且还可以看出值域
df.plot.box()
```

```python
df1 = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
df1.head()
```

```python
#7. hexbin()
df1.plot.hexbin(x='a',y='b')

#让点变大点儿
df1.plot.hexbin(x='a',y='b',gridsize = 25,cmap='coolwarm')

#the hexbins get darker when more points are there
```

```python
#8. kennel dense
df1['a'].plot.kde()

#density
df1.plot.density() #compare different densities, so cool !!
```
